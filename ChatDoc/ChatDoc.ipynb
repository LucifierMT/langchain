### 使用 多重查询 提高文档检索精确度

```python
# 优化索引 修改askAndFindFile函数

from langchain.document_loaders import Docx2txtLoader, PyPDFLoader, UnstructuredExcelLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.embeddings import DashScopeEmbeddings
from langchain.vectorstores import Chroma

from langchain_community.llms import Tongyi
from langchain.retrievers.multi_query import MultiQueryRetriever

class ChatDoc():
    def __init__(self):
        self.doc = None
        self.splitText = []
    
    # ------读取文件的函数------
    def getFile(self):
        doc = self.doc
        loader_map = {
            "docx": Docx2txtLoader,
            "pdf": PyPDFLoader,
            "xlsx": UnstructuredExcelLoader
        }
        
        file_type = doc.split(".")[-1]
        
        loader_class = loader_map.get(file_type)  # 获取对应的加载器类（而非实例）
        
        if loader_class:
            try:
                if file_type == "xlsx":
                    loader = loader_class(doc, mode="elements")  # UnstructuredExcelLoader需要特殊参数
                else:
                    loader = loader_class(doc)
                text = loader.load()
                return text
            except Exception as e:
                print(f"Error loading {file_type}: {e}")
                return None
        else:
            print(f"Unsupported file type: {file_type}")
            return None 
        
    # ------处理文件的函数------
    def splitSentences(self):
        full_text = self.getFile()
        if full_text != None:
            text_splitter = CharacterTextSplitter(
                separator="\n\n",
                chunk_size=100,
                chunk_overlap=20,
                length_function=len
            )
            texts = text_splitter.split_documents(full_text)
            self.splitText = texts

    # ------向量化与存储的函数------
    def embedAndStore(self):
        embeddings = DashScopeEmbeddings()
        db = Chroma.from_documents(self.splitText, embeddings)
        return db
    
    # ------提问并找到相关的文本块------
    def askAndFindFile(self, question):
        db = self.embedAndStore()
        # 把问题交给llm进行多角度扩展
        llm = Tongyi(temparature=0)
        retriever_from_llm = MultiQueryRetriever.from_llm(
            llm=llm,
            retriever=db.as_retriever(search_kwargs={"k": 2}),
        )
        # print(retriever_from_llm)
        return retriever_from_llm.get_relevant_documents(question)
        
# 测试代码
doc = ChatDoc()
doc.doc = "resource/test.pdf"
doc.splitSentences()
# doc.embedAndStore()
# doc.askAndFindFile("在武汉应该去哪里体检？")

import logging
logging.basicConfig(level=logging.INFO)
logging.getLogger("langchain.retriever.multi_query").setLevel(logging.INFO)

unique_doc = doc.askAndFindFile("在武汉应该去哪里体检？")
print(unique_doc)
```

### 使用 上下文压缩 降低冗余信息

```python
from langchain.document_loaders import Docx2txtLoader, PyPDFLoader, UnstructuredExcelLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.embeddings import DashScopeEmbeddings
from langchain.vectorstores import Chroma

from langchain_community.llms import Tongyi
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import LLMChainExtractor
from langchain.prompts import PromptTemplate

class ChatDoc():
    def __init__(self):
        self.doc = None
        self.splitText = []
    
    # ------读取文件的函数------
    def getFile(self):
        doc = self.doc
        loader_map = {
            "docx": Docx2txtLoader,
            "pdf": PyPDFLoader,
            "xlsx": UnstructuredExcelLoader
        }
        
        file_type = doc.split(".")[-1]
        
        loader_class = loader_map.get(file_type)  # 获取对应的加载器类（而非实例）
        
        if loader_class:
            try:
                if file_type == "xlsx":
                    loader = loader_class(doc, mode="elements")  # UnstructuredExcelLoader需要特殊参数
                else:
                    loader = loader_class(doc)
                text = loader.load()
                return text
            except Exception as e:
                print(f"Error loading {file_type}: {e}")
                return None
        else:
            print(f"Unsupported file type: {file_type}")
            return None 
        
    # ------处理文件的函数------
    def splitSentences(self):
        full_text = self.getFile()
        if full_text != None:
            text_splitter = CharacterTextSplitter(
                separator="\n\n",
                chunk_size=500,
                chunk_overlap=100,
                length_function=len
            )
            texts = text_splitter.split_documents(full_text)
            self.splitText = texts

    # ------向量化与存储的函数------
    def embedAndStore(self):
        embeddings = DashScopeEmbeddings()
        db = Chroma.from_documents(self.splitText, embeddings)
        return db
    
    # ------提问并找到相关的文本块------
    def askAndFindFile(self, question):
        db = self.embedAndStore()
        retriever = db.as_retriever(search_kwargs={"k": 2})
        llm = Tongyi(
            temperature=0,
            model_name="qwen-plus"
        )
        prompt_template = """请从以下内容中提取与问题相关的部分：
            {question}
            上下文：
            {context}
            相关提取：
        """
        prompt = PromptTemplate(
            template=prompt_template,
            input_variables=["question", "context"]
        )
        llm_chain = prompt | llm
        compresser = LLMChainExtractor(llm_chain=llm_chain)
        compresser_retriever = ContextualCompressionRetriever(
            base_retriever=retriever,
            base_compressor=compresser,
        )
        return compresser_retriever.get_relevant_documents(query=question)

        
# 测试代码
doc = ChatDoc()
doc.doc = "resource/test.pdf"
doc.splitSentences()
# doc.embedAndStore()
# doc.askAndFindFile("在武汉应该去哪里体检？")

import logging
logging.basicConfig(level=logging.INFO)
logging.getLogger("langchain.retriever.multi_query").setLevel(logging.INFO)

unique_doc = doc.askAndFindFile("在武汉应该去哪里体检？")
print(unique_doc)
```

### 和文件聊天

```python
from langchain.document_loaders import Docx2txtLoader, PyPDFLoader, UnstructuredExcelLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.embeddings import DashScopeEmbeddings
from langchain.vectorstores import Chroma

from langchain_community.llms import Tongyi
from langchain.prompts import ChatPromptTemplate

class ChatDoc():
    def __init__(self):
        self.doc = None
        self.splitText = []
        self.template = [
            ("system","你是一个处理文档的秘书，你从不说自己是一个大模型，你会根据下面提供的上下文内容来继续回答问题。"
            "上下文内容是：{context}"),
            ("user","我想问你一个问题：{question}"),
            ("assistant","好的，我会根据上下文内容来回答你的问题。")
        ]
        self.prompt = ChatPromptTemplate.from_messages(self.template)
    
    # ------读取文件的函数------
    def getFile(self):
        doc = self.doc
        loader_map = {
            "docx": Docx2txtLoader,
            "pdf": PyPDFLoader,
            "xlsx": UnstructuredExcelLoader
        }
        
        file_type = doc.split(".")[-1]
        
        loader_class = loader_map.get(file_type)  # 获取对应的加载器类（而非实例）
        
        if loader_class:
            try:
                if file_type == "xlsx":
                    loader = loader_class(doc, mode="elements")  # UnstructuredExcelLoader需要特殊参数
                else:
                    loader = loader_class(doc)
                text = loader.load()
                return text
            except Exception as e:
                print(f"Error loading {file_type}: {e}")
                return None
        else:
            print(f"Unsupported file type: {file_type}")
            return None 
        
    # ------处理文件的函数------
    def splitSentences(self):
        full_text = self.getFile()
        if full_text != None:
            text_splitter = CharacterTextSplitter(
                separator="\n\n",
                chunk_size=500,
                chunk_overlap=100,
                length_function=len
            )
            texts = text_splitter.split_documents(full_text)
            self.splitText = texts

    # ------向量化与存储的函数------
    def embedAndStore(self):
        embeddings = DashScopeEmbeddings()
        db = Chroma.from_documents(self.splitText, embeddings)
        return db
    
    # ------提问并找到相关的文本块------
    def askAndFindFile(self, question):
        db = self.embedAndStore()
        # retriever = db.as_retriever(search_type="mmr")
        retriever = db.as_retriever(search_type="similarity_score_threshold", search_kwargs={"k": 1, "score_threshold": 0.5})
        return retriever.get_relevant_documents(query=question)
    
    # ------用自然语言和文档聊天------
    def chatWithDoc(self, question):
        _content = ""
        context = self.askAndFindFile(question)
        for i in context:
            _content += i.page_content
        messages = self.prompt.format_messages(
            question=question,
            context=_content
        )
        llm = Tongyi(
            temperature=0,
            model_name="qwen-plus"
        )
        return llm.invoke(messages)


        
# 测试代码
doc = ChatDoc()
doc.doc = "resource/test.pdf"
doc.splitSentences()
doc.chatWithDoc("在武汉应该去哪里体检？")
```
