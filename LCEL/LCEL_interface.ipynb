- 8种不同的接口方式

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_community.chat_models.tongyi import ChatTongyi

llm=ChatTongyi(model="qwen-plus",max_token=200)

prompt=ChatPromptTemplate.from_template("给我讲一个关于{topic}的笑话")

chain = prompt | llm
```

#### Input schema

```python
chain.input_schema.model_json_schema()
```

```python
prompt.input_schema.model_json_schema()
```

```python
llm.input_schema.model_json_schema()
```

#### Output schema

```python
chain.output_schema.model_json_schema()
```

#### Stream

```python
for s in chain.stream({"topic":"男人和女人"}):
    print(s.content,end='',flush=True)
```

#### Invoke

```python
chain.invoke({"topic":"小孩和老人"})
```

#### Batch

```python
chain.batch(
    [
        {
            "topic":"男人"
        },
        {
            "topic":"女人"
        }
    ]
)
```

```python
# 通过 max_concurrency 控制并发数
chain.batch(
    [
        {
            "topic":"狗"
        },
        {
            "topic":"猫"
        },
        {
            "topic":"鸡"
        }
    ],
    config={
        "max_concurrency":2
    }
)
```

#### Async Stream

```python
async for s in chain.astream({"topic":"天和地"}):
    print(s.content, end='',flush=True)
```

#### Async Invoke

```python
await chain.ainvoke({"topic":"秦始皇"})
```

#### Async Batch

```python
await chain.abatch(
    [
        {
            "topic":"狗"
        },
        {
            "topic":"猫"
        }
    ]
)
```

#### 异步获取中间步骤

```python

from langchain_community.vectorstores import FAISS
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_community.chat_models.tongyi import ChatTongyi
from langchain_community.embeddings import DashScopeEmbeddings

vectorstore=FAISS.from_texts(
    ["兜兜专干老婆，晨晨专干兜兜"],
    embedding=DashScopeEmbeddings()
)

retriever=vectorstore.as_retriever()

template='''根据上下文回答问题：{context}\n问题：{question}'''

prompt=ChatPromptTemplate.from_template(template)

llm=ChatTongyi(model="qwen-plus",temperature=0.2,max_token=1000)

chain=(
    {
        "context":retriever.with_config(run_name="Docs"),
        "question":RunnablePassthrough()
    }
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
async for chunk in chain.astream_log(
    "说一下这里的人物克制关系",
    include_names=["Docs"]
):
    print("-"*10)
    print(chunk)
```

#### 并行支持

```python
from langchain_core.runnables import RunnableParallel

chain1=ChatPromptTemplate.from_template("给我讲一个关于{topic}的笑话") | llm
chain2=ChatPromptTemplate.from_template("给我写一首关于{topic}的诗歌") | llm

combined=RunnableParallel(joke=chain1,poem=chain2)
```

```python
%%time
chain1.invoke({"topic":"特朗普"})
```

```python
%%time
chain2.invoke({"topic":"周杰伦"})
```

```python
%%time
# 并行处理
combined.invoke(
    [
        {
            "topic":"美国总统"
        }
    ]
)
```

```python
%%time
# 并行批处理
combined.batch(
    [
        {
            "topic":"美国总统"
        },
        {
            "topic":"周杰伦"
        }
    ]
)
```
